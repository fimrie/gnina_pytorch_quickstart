```python
def prepare_protein_openbabel(input_pdb, output_pdbqt):
    mol = next(pybel.readfile("pdb", input_pdb))
    mol.addh()  # Add hydrogens (default = neutral pH)
    try:
        mol.calccharges(model="gasteiger")  # Limited value for proteins
    except:
        pass
    mol.write("pdbqt", output_pdbqt, overwrite=True)

for pdb_id in pdb_ids:
    prepare_protein_openbabel(f'./data/raw_data/CASF-2016/{pdb_id}/{pdb_id}_pocket.pdb', f'./data/raw_data/CASF-2016/{pdb_id}/{pdb_id}_pocket_prepared.pdbqt')
```

```python
def prepare_ligand_openbabel(input_sdf, output_pdbqt):
    mol = next(pybel.readfile("sdf", input_sdf))
    mol.addh()
    mol.calccharges(model="gasteiger")
    mol.write("pdbqt", output_pdbqt, overwrite=True)

for pdb_id in pdb_ids:
    prepare_ligand_openbabel(f'./data/raw_data/CASF-2016/{pdb_id}/{pdb_id}_ligand.sdf', f'./data/raw_data/CASF-2016/{pdb_id}/{pdb_id}_ligand_prepared.pdbqt')
```

```python
import subprocess
import os
import sys
from pathlib import Path

def supress_stdout(func):
    def wrapper(*a, **ka):
        with open(os.devnull, 'w') as devnull:
            with contextlib.redirect_stdout(devnull):
                return func(*a, **ka)
    return wrapper


class PrepLig(object):
    def __init__(self, input_mol, mol_format):
        if mol_format == 'sdf': 
            self.ob_mol = next(pybel.readfile(mol_format, input_mol))
        else:
            raise ValueError(f'mol_format {mol_format} not supported')
        
    def addH(self, polaronly=False, correctforph=True, PH=7): 
        self.ob_mol.OBMol.AddHydrogens(polaronly, correctforph, PH)
        obutils.writeMolecule(self.ob_mol.OBMol, 'tmp_h.sdf')

    @supress_stdout
    def get_pdbqt(self, lig_pdbqt=None):
        preparator = MoleculePreparation()
        self.ob_mol.OBMol.AddHydrogens(False, True, 7.4)
        preparator.prepare(self.ob_mol.OBMol)
        if lig_pdbqt is not None: 
            preparator.write_pdbqt_file(lig_pdbqt)
            return 
        else: 
            return preparator.write_pdbqt_string()
        
from parmed import load_file

def remove_hetatoms_and_waters(pdb_in: str, pdb_out: str):
    with open(pdb_in, 'r') as f_in, open(pdb_out, 'w') as f_out:
        for line in f_in:
            if line.startswith("HETATM"):
                continue
            else:
                f_out.write(line)

class PrepProt(object): 
    def __init__(self, pdb_file): 
        self.prot = pdb_file
    
    def del_water(self, dry_pdb_file): # optional
        with open(self.prot) as f: 
            lines = [l for l in f.readlines() if l.startswith('ATOM') or l.startswith('HETATM')] 
            dry_lines = [l for l in lines if not 'HOH' in l]
        
        with open(dry_pdb_file, 'w') as f:
            f.write(''.join(dry_lines))
        self.prot = dry_pdb_file
        
    #def addH(self, prot_pqr):  # call pdb2pqr
    #    self.prot_pqr = prot_pqr
    #    subprocess.Popen(['pdb2pqr30','--ff=AMBER', self.prot, self.prot_pqr],
    #                     stderr=subprocess.DEVNULL, stdout=subprocess.DEVNULL).communicate()

    def run_leap(self, pdb_file: str, output_prefix: str):
        leap_input = f"""
source leaprc.protein.ff14SB
protein = loadpdb "{pdb_file}"
check protein
saveamberparm protein {output_prefix}.prmtop {output_prefix}.inpcrd
savepdb protein {output_prefix}_ff14sb.pdb
quit
"""
        with open("leap.in", "w") as f:
            f.write(leap_input)
    
        result = subprocess.run(
            ["tleap", "-f", "leap.in"],
            stdout=subprocess.DEVNULL, stderr=subprocess.PIPE, text=True
        )
        if result.returncode != 0:
            #print(result.stdout)
            print("LEaP failed:\n", result.stderr)
            raise RuntimeError("LEaP failed")
        #structure = load_file(f"{output_prefix}.prmtop", f"{output_prefix}.inpcrd")
        #structure.save(f"{output_prefix}.mol2", format="mol2", overwrite=True)

    def convert_to_pqr(self, prmtop: str, inpcrd: str, output_pqr: str):
        structure = load_file(prmtop, inpcrd)
        structure.save(output_pqr, format="pqr", overwrite=True)
        
    def addH(self, base, prot_pqr):
        self.prot_pqr = prot_pqr
        #print(base, f'{base}.pqr')

        remove_hetatoms_and_waters(self.prot, f"{base}_nohetam.pdb")
        res=subprocess.run(["pdb4amber", "-i", f"{base}_nohetam.pdb", "-o", f"{base}_clean.pdb", "--dry", "--nohyd", "--reduce"],
                       stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL
                      )
        #print(res.stdout)
        
        self.run_leap(f"{base}_clean.pdb", base)
        self.convert_to_pqr(f"{base}.prmtop", f"{base}.inpcrd", f"{base}.pqr")
    
    def get_pdbqt(self, base, prot_pdbqt):
        prepare_receptor = os.path.join(AutoDockTools.__path__[0], 'Utilities24/prepare_receptor4.py')
        subprocess.Popen(['python3', prepare_receptor, '-r', self.prot_pqr, '-o', prot_pdbqt, '-C', '-U', 'waters_nonstdres'],
                         stderr=subprocess.DEVNULL, stdout=subprocess.DEVNULL).communicate()
```
